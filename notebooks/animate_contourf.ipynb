```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation,rc
rc('animation', html='html5')
from IPython.display import HTML
%matplotlib inline
import k2movie
mov=k2movie.movie(name='C/2013 A1',
                  outfile='comet.mp4',
                  channel=None,
                  cadence=np.arange(2500, 3800,10),
                  campaign=2,
                  cmap='inferno',
                  tol=50,
                  inset=False,
                  verbose=False,
                  data_dir='../database/',
                  output_dir='../movies/',
                vmin=1,vmax=3.5,return_radec=True)

#fig,ax=plt.subplots(figsize=(10,10))
#fig=mov.inspect(ax,radec=True)
ar,ra,dec=mov.compute(return_ar=True)
```

```python
i=0
plt.contourf(ra.T[:,:,i],dec.T[:,:,i],np.log10(ar[:,:,i]))
```
